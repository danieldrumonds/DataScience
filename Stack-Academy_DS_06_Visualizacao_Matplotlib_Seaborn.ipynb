```python
# Importa o numpy
import numpy as np
```

```python
%matplotlib inline

# Quantidade de vendas para o Produto A
valores_produto_A = [6,7,8,4,4]

# Quantidade de vendas para o Produto B
valores_produto_B = [3,12,3,4.1,6]

# Cria eixo x para produto A e produto B com uma separação de 0.25 entre as barras
x1 =  np.arange(len(valores_produto_A))
x2 = [x + 0.25 for x in x1]

# Plota as barras
plt.bar(x1, valores_produto_A, width=0.25, label = 'Produto A', color = 'b')
plt.bar(x2, valores_produto_B, width=0.25, label = 'Produto B', color = 'y')

meses = ['Agosto','Setembro','Outubro','Novembro','Dezembro']
plt.xticks([x + 0.25 for x in range(len(valores_produto_A))], meses)

plt.legend()

plt.title("Quantidade de Vendas")
plt.show()
```

# 6.8 - Matplotlib: Configurando Eixos e Subplots

**É possível customizar a largura ou altura do gráfico para acomodar melhor seus dados.**

*  Exemplo: gráficos de dados financeitos em uma granularidade muito baixa (minutos)

```python
x = [1,2,3,4]
y = [10,20,25,40]

# Altera a largura e altura do gráfico (width,height) 
fig, ax = plt.subplots(figsize=(20,5))

# Plota o gráfico
ax.plot(x,y, 'r--')
```

**Plotando dois gráficos (eixos) na mesma figura**

```python
# Define as configurações dos plots
# Cada plot terá o mesmo tamanho de figuras

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Dados para cada subplot
ax1.bar([1,2,3],[3,4,5])

ax2.barh([0.5,1,2.5],[0,1,2])

plt.show()
```

**Define configuracões dos eixos com o método "set()"**

```python
import matplotlib.pyplot as plt

# Define as configurações dos plots
# Cada plot terá o mesmo tamanho de figuras
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

# Dados para cada subplot
ax1.bar([1,2,3],[3,4,5], color='r')

ax2.barh([0.5,1,2.5],[0,1,2])

# Usa o método set para definir configurações do  gráfico
ax1.set(title="Gráfico de Barras Verticais", xlabel="Eixo x", ylabel=" Eixo y")

ax2.set(title="Gráfico de Barras Horizontais", xlabel="Eixo x", ylabel="Eixo y")

plt.show()
```

## Gráficos com Regressão

* Quando temos muitas variáveis quantitativas em nossos dados é interessante visualizar como estas se relacionam.
* Podemos visualizar essas informações com linhas de regressão.
* Com modelos de regressão simples é possível checar se existe alguma **correlacão** entre algumas variáveis.

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

**Exibe a linha de regressão para visualizar correlação entre as variáveis**

```python
%matplotlib inline
sns.lmplot(x="total_bill", y="tip", hue="sex", palette="Set1", data=tips);
```

**Usando o parâmetro col para segregar os gráficos pelo valor da coluna categórica *time***

```python
%matplotlib inline
sns.lmplot(x="total_bill",y="tip", hue="smoker", col="time", palette="Set1", data=tips);
```

**Parâmetros scatter_kws e line_kws**

```python
%matplotlib inline

sns.lmplot(x="total_bill",
           y="tip", 
           hue="smoker", 
           col="time", 
           palette="Set1",
           data=tips, 
           scatter_kws={'s': 90, 'alpha': 0.5}, #s=size alpha=transparencia
           line_kws={'lw': 4}); # line_kws=largura da linha
```

# 6.24 - Seaborn: Aplicando Jointplot e Hexbin

## Jointplot (ScatterPlot e Histograma)

* Quando queremos ver a distribuição de duas variáveis podemos usar o jointplot.
* União de gráficos do tipo Scatter Plot e Histogramas para as duas variáveis.
* Mostra o Scatter plot para as duas variáveis e um histograma para cada variável separadamente.

```python
sns.jointplot(x="tip", y="total_bill", color='r', data=tips);
```

* Adequado para visualização de grande conjuntos de dados.
* Tende a funcionar melhor com as cores em branco de fundo

```python
sns.jointplot(x="tip", y="total_bill", color='r', kind="hex", data=tips)
```

```python
sns.jointplot(x="tip", y="total_bill", color='b', kind="kde", data=tips)
```
