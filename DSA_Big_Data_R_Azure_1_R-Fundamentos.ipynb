```r
# Else
if (x < 7) {
  "Este número é menor que 7"
} else {
  "Este número não é menor que 7"
}
```

```r
# Comandos podem ser ainhados
x = 7
if (x < 7) {
  "Este número é menor que 7"
} else if(x == 7) {
  "Este é o número 7"
} else {
  "Este número não é menor que 7"
}
```

```r
# Funções Built-in - Não tente recriar a roda
# Comparação de eficiência entre função vetorizada e função "vetorizada no R"

x <- 1:1000000
```

```r
# função que calcula a raiz quadrada de cada elemento de um vetor de números
meu_sqrt <- function(numeros) {
  resp <- numeric(length(numeros))
  for(i in seq_along(numeros)) {
    resp[i] <- sqrt(numeros[i])
  }
  return(resp)
}
```

```r
#system.time compara tempo de execução
system.time(x2a <- meu_sqrt(x))

system.time(x2b <- sqrt(x))

identical(x2a, x2b)
```

# Capítulo 3.5: A família Apply - uma forma elegante de fazer loops

- apply() - arrays e matrizes
- tapply() - os vetores podem ser divididos em diferentes subsets
- lapply() - vetores e listas
- sapply() - versão amigável da lapply
- vapply() - similar a sapply, com valor de retorno modificado
- rapply() - similar a lapply()
- eapply() - gera uma lista
- mapply() - similar a sapply, multivariada
- by

#### Se você estiver trabalhando com os objetos:

list, numeric, character (list/vecor) => sapply ou lapply<BR>
matrix, data.frame (agregação por coluna) => by / tapply<BR>
Operações por linha ou operações específicas => apply<BR>

# Capítulo 3.8: Expressões Regulares

**Expressões Regulares:** Recurso usado para verificar se existe um padrão em uma string ou vetor de caracteres

**grepl()** Retorna TRUE quando um padrão é encontrado<BR>
**grep** Retorna um vetor de Índices dos caracteres que contam o padrão especificado<BR>
**sub()** Substitui o primeiro caracter encontrado de acordo com o padrão especificado<BR>
**gsub()** Substitui todos os caracteres encontrados de acordo com o padrão especificado

grep(pattern, x, ignore.case = FALSE, perl = FALSE, value = FALSE, fixed = FALSE, useBytes = FALSE, invert = FALSE)<BR>
grepl(pattern, x, ignore.case = FALSE, perl = FALSE, fixed = FALSE, useBytes = FALSE)<BR>
sub(pattern, replacement, x, ignore.case = FALSE, perl = FALSE, fixed = FALSE, useBytes = FALSE)<BR>
gsub(pattern, replacement, x, ignore.case = FALSE, perl = FALSE, fixed = FALSE, useBytes = FALSE)<BR>
regexpr(pattern, text, ignore.case = FALSE, perl = FALSE, fixed = FALSE, useBytes = FALSE)<BR>
gregexpr(pattern, text, ignore.case = FALSE, perl = FALSE, fixed = FALSE, useBytes = FALSE)

```r
# Lendo o arquivo
arquivo <- "http://www.sthda.com/sthda/RDoc/example-files/martin-luther-king-i-have-a-dream-speech.txt"
texto <- readLines(arquivo)
```

```r
# Carregando os dados como Corpus
docs <- Corpus(VectorSource(texto))
```

```r
# Pré-processamento
inspect(docs) # Verificar metadados
```

```r
toSpace <- content_transformer(function (x , pattern ) gsub(pattern, " ", x))
docs <- tm_map(docs, toSpace, "/") # Busca pelo padrão e converte em espaço
docs <- tm_map(docs, toSpace, "@")
docs <- tm_map(docs, toSpace, "\\|")
```

```r
# Converte o texto para minúsculo
docs <- tm_map(docs, content_transformer(tolower))
```

```r
# Remove números
docs <- tm_map(docs, removeNumbers)
```

```r
# Remove as palavras mais comuns do idioma inglês
docs <- tm_map(docs, removeWords, stopwords("english"))
```

```r
# Você pode definir um vetor de palavras (stopwords) a serem removidas do texto
docs <- tm_map(docs, removeWords, c("blabla1", "blabla2"))
```

```r
# Remove pontuação
docs <- tm_map(docs, removePunctuation)
```

```r
# Elimina espaços extras
docs <- tm_map(docs, stripWhitespace)
```

```r
# Text stemming: Juntar palavras que aparecem com muita frequência
docs <- tm_map(docs, stemDocument)
```

```r
# Converte em uma matriz
dtm <- TermDocumentMatrix(docs)
m <- as.matrix(dtm)
v <- sort(rowSums(m),decreasing=TRUE)
d <- data.frame(word = names(v),freq=v)
```

```r
# Dataset final com as palavras que mais aparecem e a frequência
head(d, 10)
```
